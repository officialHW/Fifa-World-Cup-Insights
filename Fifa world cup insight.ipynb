# Handling missing values

#### we can handle missing values by 
#### 1. drop the missing values: its the easiest but not always teh most convinient as you might possibly be dropping other datas if they are more than the total number of rows you have in the dataset
#### 2. imputation: replacing the missing values with something, we have 2 ways of imputation
####    - simple imputation: here we are saying where there is a null fill it with ***, here we are determining the input if there are no values in it
####    - statistical imputation: here we basically use a statistical apporach to filling in the missing value such as mean or mode of the data.
####    -  advanced imputation : talking about building models that could generate an accurate data that fits the place of that missing data
####    -  back fill and forward fill: is essentially saying replace the missing value with the last known  value or the next known value

### note -
#### there is no standard way of handling missing values, its always a discussion with stakeholders to know what is best to serve for the dataset

```python
# add a new column
# Create a new column named 'Total Goals' that contains the sum of home and away goals scored in each match.

df2['Total Goals'] =df2['Home Goals'] + df2['Away Goals']

df2
```

```python
# add conditional column
# Create a conditional column â€˜Score Categoryâ€™ that contains over â€˜High Scoringâ€™ for matches with more than 2 goals and â€˜Low Scoringâ€™ otherwise

df2['Score Category'] = np.where(df2['Total Goals'] > 2, 'High Scoring', 'Low Scoring')

df2
```

```python
## method 2: using a custom or lambda function
def check_score_category(x):
    if x > 2:
        return 'High Scoring'
    else:
        return 'Low Scoring'
```

```python
## using custom function
df2['Score Category2'] = df2['Total Goals'].apply(check_score_category)
df2
```

```python
#3. How many goals has each team scored in the FIFA World Cup? Which team has scored 

# approach is to get all the goals scored by each country as home team; 
# get all goals scored by each country as away team;
# concatenate (combine) the two arrays from step 1 and 2
# then get the total goals for each country
```

```python
# get total goals as home team and as away team for each country
home_goals_df = df2.groupby('Home Team')['Home Goals'].sum().reset_index()
away_goals_df = df2.groupby('Away Team')['Away Goals'].sum().reset_index()
```

```python
home_goals_df
```

```python
away_goals_df
```

```python
# combine the two arrays/dataframes
total_goals_df = pd.concat([home_goals_df.rename(columns={'Home Team': 'Team', 'Home Goals': 'Goals'}),
                            away_goals_df.rename(columns={'Away Team': 'Team', 'Away Goals': 'Goals'})],
                           ignore_index=True)

# group by country to get total goal by each country
total_goals_by_team = total_goals_df.groupby('Team')['Goals'].sum().sort_values(ascending=False)
```

```python
#5. Goals per game ratio for each team

## approach
#1. get the total goals scored by each team (this is already done in previous question)
#2. get the total games played by each team (get away games and home games and add for each team)
#3. calculate the ratio
```

```python
# get total games as home team and as away team for each country
home_games = df2['Home Team'].value_counts()
away_games = df2['Away Team'].value_counts()

# add the two series
total_games = home_games.add(away_games, fill_value=0)  
## fill_value is used to handle any case of missing key ie country in home_games but not in away games
```

```python
total_goals_by_team
```

```python
total_games
```

```python
#calculate the goal per game ratio
goals_per_game_ratio = total_goals_by_team/total_games
goals_per_game_ratio = goals_per_game_ratio.sort_values(ascending=False)
goals_per_game_ratio.round(2)   #round is to round up or down decimal places
```

```python
## what percentage of games went to Extra time in the knockout stage in 2010

#approach
#1. filter the dataframe for stage != group stage and year = 2010
#2. count the total numer of games
#3. filter further for win condition != normal time
#4. count the total number of games
#5. calculate the ratio
```

```python
#filter further for only games that went to extra time
condition = df_2010_knockout['Win Conditions'] != 'NaN'
df_2010_knockout_ET = df_2010_knockout[condition]
df_2010_knockout_ET
```

```python
#get the count of games
total_count_ET = len(df_2010_knockout_ET)

# get the ratio
total_count_ET / total_count
```
